```python
# Imports
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Load data
df = pd.read_csv('https://query.data.world/s/5yzzglbcb2nabj2rmoqjacf5os5oes')
```

```python
# Filters
regular_season = df['Season type'] == 'Regular Season'
playoffs = df['Season type'] == 'Playoffs'

# Dataframe of the filters
df_rs = df.loc[regular_season, :]
df_po = df.loc[playoffs, :]
```

```python
# Create a stacked bar chart for playoff fouls
po_fouls = df_po.groupby(['Season'])[['Shooting', 'Personal', 'Loose ball', 'Offensive charge', 'Technical']].sum()
po_fouls.plot(kind = 'bar', stacked = True, figsize = (12,8))

# Set labels, title, legend and font size
plt.xlabel('Season', fontsize = 18)
plt.ylabel('Number of Fouls', fontsize = 18)
plt.xticks(fontsize = 15, rotation = 0)
plt.yticks(fontsize = 15)
plt.title('Fouls in the NBA Playoffs', fontsize = 20)
plt.legend(bbox_to_anchor = (1, 1), fontsize = 15)

# Create a stacked bar chart for regular season fouls
rs_fouls = df_rs.groupby(['Season'])[['Shooting', 'Personal', 'Loose ball', 'Offensive charge', 'Technical']].sum()
rs_fouls.plot(kind = 'bar', stacked = True, figsize = (12,8))

# Set labels, title, legend and font size
plt.xlabel('Season', fontsize = 18)
plt.ylabel('Number of Fouls', fontsize = 18)
plt.xticks(fontsize = 15, rotation = 0)
plt.yticks(fontsize = 15)
plt.title('Fouls in the NBA Regular Season', fontsize = 20)
plt.legend(bbox_to_anchor = (1, 1), fontsize = 15)
;
```
