```python
# Imports
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Load data
df = pd.read_csv('https://query.data.world/s/5yzzglbcb2nabj2rmoqjacf5os5oes')
```

```python
# Filters
season_1617 = df['Season'] == '2016-17'
season_1718 = df['Season'] == '2017-18'
season_1819 = df['Season'] == '2018-19'
season_1920 = df['Season'] == '2019-20'
regular_season = df['Season type'] == 'Regular Season'
playoffs = df['Season type'] == 'Playoffs'

# Dataframe of the filters
df_s1617_rs = df.loc[season_1617 & regular_season, :]
df_s1617_po = df.loc[season_1617 & playoffs, :]
df_s1718_rs = df.loc[season_1718 & regular_season, :]
df_s1718_po = df.loc[season_1718 & playoffs, :]
df_s1819_rs = df.loc[season_1819 & regular_season, :]
df_s1819_po = df.loc[season_1819 & playoffs, :]
df_s1920_rs = df.loc[season_1920 & regular_season, :]
df_s1920_po = df.loc[season_1920 & playoffs, :]
df_rs = df.loc[regular_season, :]
df_po = df.loc[playoffs, :]
```

```python
# Set figure size
plt.figure(figsize = (50, 10))


# Lists of data
season = ['2016-17', '2017-18', '2018-19', '2019-20']
po_fouls = [df_s1617_po['Total fouls'].sum(), df_s1718_po['Total fouls'].sum(), df_s1819_po['Total fouls'].sum(), df_s1920_po['Total fouls'].sum()]
rs_fouls = [df_s1617_rs['Total fouls'].sum(), df_s1718_rs['Total fouls'].sum(), df_s1819_rs['Total fouls'].sum(), df_s1920_rs['Total fouls'].sum()]

# Bar graph of total fouls in the playoffs
plt.subplot(131)
plt.bar(season, po_fouls)

# Set labels, title and font size
plt.xlabel('Season', fontsize = 18)
plt.ylabel('Total Fouls', fontsize = 18)
plt.xticks(fontsize = 15)
plt.yticks(fontsize = 15)
plt.title('Total Fouls in the Playoffs', fontsize = 20)

# Bar graph of total fouls in the regular season
plt.subplot(132)
plt.bar(season, rs_fouls)

# Set labels, title and font size
plt.xlabel('Season', fontsize = 18)
plt.ylabel('Total Fouls', fontsize = 18)
plt.xticks(fontsize = 15)
plt.yticks(fontsize = 15)
plt.title('Total Fouls in the Regular Season', fontsize = 20)

;
```
